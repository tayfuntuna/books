```python
def get_authors(sure=1, ayet=1):
    url =  f"https://www.kuranmeali.com/AyetKarsilastirma.php?sure={sure}&ayet={ayet}"
    page = requests.get(url, allow_redirects=False)
    soup = BeautifulSoup(page.content, "html.parser")
    divs = soup.find('div', attrs={'id':'Mealler'}).find_all("div")
    d = {}
    for di in divs:
        if 'id' in di.attrs: 
               
                d[di.attrs['id']] = di.find('div').text if  di.find('div') else  di.text
    return d
authors = get_authors(sure=1, ayet=1)
```

```python
authors
```

```python
def get_ayet(sure, ayet):
    url =  f"https://www.kuranmeali.com/AyetKarsilastirma.php?sure={sure}&ayet={ayet}"
    page = requests.get(url, allow_redirects=False)
    soup = BeautifulSoup(page.content, "html.parser")
    divs = soup.find('div', attrs={'id':'Mealler'}).find_all("div")
   
     
    
    data = []
    for di in divs:
        if 'id' in di.attrs: 
                d = {}
                d['meal'] = di.attrs['id']
                d['sure'] = sure
                d['ayet'] = ayet
                d['text'] = di.find('span').text if  di.find('span') else  di.text
                data.append(d)
    next_url = soup.find('a', attrs={'id':'ileri'})
    if next_url:
             
        sure,ayet=next_url.attrs["href"].replace("?sure=",'').replace('ayet=','').split("&")
    else:
        sure = sure+1
        ayet =1
    return data,int(sure),int(ayet)
```

```python
authors = pd.DataFrame([authors])
authors.to_csv("authors", index=False)
```

```python
all_data = []
sure, ayet = 1,1
```

```python

counter = 0
while sure<116:
    counter+=1
    data, sure, ayet= get_ayet(sure, ayet)
    print(sure,ayet)
    all_data.extend(data)
    if counter%101==0:
        sleep(1)
```
